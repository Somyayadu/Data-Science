```python
import numpy as np
import matplotlib .pyplot as plt 
import pandas as pd
import seaborn as sns


```

```python
arr = np.random. randint (200,40000,100)
arr
```

```python
# lineplot 

arr.sort()
plt.plot(arr)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[19,20,37,40,57,60]
plt.plot(x,y,marker="s",ms=15,mfc="green",mec="red",color="black",lw=3,ls="--",alpha=0.5,label="lineplot")               
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()

# ms -- marker size
# mfc -- marker face colour
# mec -- marker edge colour 
# color -- line colour 
# lw -- line weights
# ls -- line shape
# alpha -- transparancy
```

```python
#scattrplot
plt.scatter(x,y)
plt.show()
```

```python
x=["mp","up","rj","tn","mh"]
y=[10,20,30,40,50]
colours=["red","green","blue","yellow","orange"]
hatch_patterns=["/","\\","|","-","*"]
plt.bar(x,y,color=colours,hatch=hatch_patterns,label=x,width=0.3 )  #bar plot
#plt.barh(x,y)
#plt.grid()
plt.legend()
plt.show()
```

```python
## histogram -- distribution n of data in interval(bins)

arr=np.random.randint(20000,40000,100)
plt.hist(arr,bins=30)
plt.show()
```

```python
## piechart -- percentage

x=["mp","up","rj","gj"]
y=[10,20,30,40]
explodes =(0.1,0,0,0)
plt.pie(y,labels=x,autopct="%1.2f%%",shadow=True,explode=explodes)
plt.savefig("piechart.png")
plt.legend()
plt.show()
```

```python
##subplot

x =np.linspace(0,10,100)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)            ## 2rows ,2columns , 1graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()
plt.show()


```
